```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# load Data
path='clean_usedcars.csv'

df = pd.read_csv(path)
df.head(3)
```

## Multiple Linear Regression
### Fit
<p>From the previous section (data_analysis.ipynb)  we know that other good predictors of price could be:</p>

<ul>
    <li>Curb-weight</li>
    <li>Engine-size</li>
    <li>Horsepower</li>
    <li>City-mpg</li>
    <li>Highway-mpg</li>
</ul>

```python
lm = LinearRegression()
```

```python
Z = df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg', 'city-mpg']]
lm.fit(Z, df['price'])
```

```python
lm1 = LinearRegression()
Z1 = df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']]
lm1.fit(Z1, df['price'])
```

```python
lm2 = LinearRegression()
Z2 = df[['horsepower', 'curb-weight', 'engine-size']]
lm2.fit(Z2, df['price'])
```

```python
lm3 = LinearRegression()
Z3 = df[['curb-weight', 'engine-size']]
lm3.fit(Z3, df['price'])
```

## In-Sample Evaluation

### How to evaluate a linear model
when it is about a linear regresion, we can evaluate by visualasing by using a regession plot or a residual plot.


```python
import seaborn as sns
```

```python
# residual plot
width = 12
height = 10
plt.figure(figsize=(width, height))
sns.residplot(x=df['engine-size'], y=df['price'])
plt.show()
```

<b> about the plot</b>
<p>We can see from this residual plot that the residuals are not randomly spread around the x-axis, leading us to believe that maybe a non-linear model is more appropriate for this data.</p>

### Evaluate multiple linear regression
We can evlation a multiple linear regression model by using  a distribution plot. We can look at the distribution of the fitted values that result from the model and compare it to the distribution of the actual values.

```python
Y_hat = lm.predict(Z)
plt.figure(figsize=(8, 6))


ax1 = sns.distplot(df['price'], hist=False, color="r", label="Actual Value")
sns.distplot(Y_hat, hist=False, color="b", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Cars')
plt.legend()

plt.show()
plt.close()
```

```python
Y_hat = lm1.predict(Z1)
plt.figure(figsize=(8, 6))


ax1 = sns.distplot(df['price'], hist=False, color="r", label="Actual Value")
sns.distplot(Y_hat, hist=False, color="b", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Cars')
plt.legend()

plt.show()
plt.close()
```

```python
Y_hat = lm2.predict(Z2)
plt.figure(figsize=(8, 6))


ax1 = sns.distplot(df['price'], hist=False, color="r", label="Actual Value")
sns.distplot(Y_hat, hist=False, color="b", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Cars')
plt.legend()

plt.show()
plt.close()
```

```python
Y_hat = lm3.predict(Z3)
plt.figure(figsize=(8, 6))


ax1 = sns.distplot(df['price'], hist=False, color="r", label="Actual Value")
sns.distplot(Y_hat, hist=False, color="b", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Cars')
plt.legend()

plt.show()
plt.close()
```

<p> visualization dont give use much informations about which models is the best. We can evlaute the models by using quantitavies measure such as  R-squared or Mean Squared Error  </p>

```python

# Find the R^2
print('lm: ', lm.score(Z, df['price']))
print('lm1: ', lm1.score(Z1, df['price']))
print('lm2: ', lm2.score(Z2, df['price']))
print('lm3: ', lm3.score(Z3, df['price']))
```

<p> according to R^2 m is the best model. We can say that ~81.593 % of the variation of price is explained by this multiple linear regression "multi_fit".</p>
